# PySCNet: A tool for reconstructing and analyzing gene regulatory network from single-cell RNA-Seq data

PySCNet includes four modules:

1) **Pro-precessing**: initialize a gnetData object consisting of Expression Matrix, Cell Attributes, Gene Attributes and Network Attributes; <br/>
2) **BuildNet**: reconstruct GRNs by various methods implemented in docker;<br/>
3) **NetEnrich**: network analysis including consensus network detection, gene module identification and trigger path prediction as well as network fusion;<br/>
4) **Visulization**: network illustration.<br/>

This tutorial guides you how to build cell specific gene regulatory network in different tools and predict consensus network, build gene modules in a stepwise manner.

```python
from __future__ import absolute_import
import warnings
warnings.filterwarnings("ignore")

import sys
import os
import itertools

from pyvis.network import Network
import pandas as pd
import copy
import numpy as np
from pyscnet.Preprocessing import gnetdata
from pyscnet.BuildNet import gne_dockercaller as gdocker
from pyscnet.NetEnrich import graph_toolkit as gt
from pyscnet.Plotting import show_net as sn


input_path = '/home/mwu/MING_V9T/PhD_Pro/PySCNet/data/'
output_path = '/home/mwu/MING_V9T/PhD_Pro/PySCNet/tutorial/result/'
```

Data can be downloaded from our published paper [GSE137007](https://www.ncbi.nlm.nih.gov/geo/query/acc.cgi?acc=GSE137007)


```python
#Import Dataset
Expr = pd.read_csv(input_path + 'Run36_KK_UMI_Top2000_Raw.txt', delimiter=' ', index_col=0)
Cell_Info = pd.read_csv(input_path + 'Run36_KK_Cell_ClusterID.txt', delimiter=' ')

Expr.columns = [x.replace('.', '-') for x in Expr.columns]

#Import Mms transcription factor
Mms_factors = pd.read_csv('../Mus_TF_and_TFcofactor/Mus_musculus_TF.txt', sep='\t')
```

```python
Cell_Info.head()
```

There are 2000 genes in the input matrix, we would like to firstly checked the regulation relationship among the transcription factors and known signatrues. To do so, we preselect the gene pool based on the prior knowledge.

```python
selected_genes = ['Tbx21', 'Cx3cr1', 'Klf2', 'S1pr1', 'S1pr4', 'Jak', 'Il18r1', 'Il18rap', 'Klrk1',
                  'Ctsd', 'Lef1', 'Klrc1', 'Klrd1', 'Lgals1', 'Lgals3', 'Vim', 'Pdcd1', 'Cd160', 'Cd244',
                  'Entpd1', 'Cd37', 'Tnfrsf9', 'Tox', 'Nr4a2', 'Psmb8', 'Ctsb', 'Lax1', 'Gzmb', 'Srgn',
                  'Tnfrsf9', 'Tnfrsf4', 'Tnfrsf18', 'Cd200r1', 'Ifngr1', 'Il2rb', 'Sh2d1a', 'Ccl3', 'Ccl4',
                  'Lsp1', 'Vasp', 'Sin3b', 'Ctsb', 'Ctsc', 'Calm1', 'Stat3', 'Id2', 'Ly6a', 'Ly6e', 'Trim30a',
                  'Il10rb', 'Fyb', 'Tnfsf8', 'Ptpn6', 'Cd84', 'Emb', 'Traf1', 'Cd9', 'Icos', 'Cd72', 'Ptpn11',
                  'Cd69', 'Sh2d1a', 'Penk', 'Slamf6', 'Dgka', 'Inpp4b', 'Tespa1', 'Ifnar2', 'Cxcr3', 'Itgb1',
                  'Prf1', 'Fasl', 'Cd244', 'Chn2', 'Fgl2', 'Tcf7', 'Gzma', 'Gzmb', 'Gzmk', 'Fasl', 'Id2', 'Tbx21', 'Pdcd1',
                  'Cd244', 'Cd160', 'Entpd1', 'Klf2', 'S1pr5', 'S1pr1', 'Cx3cr1','Tox', 'Id3', 'Top2a', 'Birc5', 'Cxcr5', 'Emoes', 'Havcr2',
            'Bcl2l11', 'Cd200', 'Siah2' ]

selected_Expr = Expr[Expr.index.isin(set(selected_genes))]
selected_Expr.shape
```

```python
#create gene information dataframe
Mms_factors['Symbol'] = [x.upper() for x in list(Mms_factors['Symbol'])]
gene_info = pd.DataFrame({'Gene': selected_Expr.index,
                          'TF_Gene': ['TF' if x.upper() in list(Mms_factors['Symbol']) else 'Gene' for x in
                                      selected_Expr.index]})
```

Next, we create gnetData object by importing gene count matrix, cell information and gene information.

```python
gne = gnetdata.Gnetdata(ExpMatrix=selected_Expr)
gne.CellAttrs = Cell_Info
gne.GeneAttrs = gene_info

gne.get_attr
```

pyscnet provides five methods for gene regulatory network reconstruction, as reported in [GRN benchmark paper](https://pubmed.ncbi.nlm.nih.gov/31907445/), GENIE3, GRNBOOST2 and PIDC were reported as competitive tools among existing GRN tools. 

```python
gne_genie3 = gdocker.rundocker(gne.deepcopy, method='GENIE3')
gne_pidc = gdocker.rundocker(gne.deepcopy, method='PIDC')
gne_grn = gdocker.rundocker(gne.deepcopy, method='GRNBOOST2')
```

For given 3 GRNs from different methods, we could do the following: 1) predict consensus network; 2) detect gene modules; 3) explore the consensus GRN via supervised random walk

```python
gne_genie3 = gdocker.buildnet(gne_genie3, top=300)
gne_genie3 = gt.get_centrality(gne_genie3)
gne_genie3 = gt.community_detect(gne_genie3)
```

 Network visualization: Export network as html or pdf

```python
sn.dynamic_netShow(gne_genie3, output_path + 'test.html')
```
